**Data**: The International Stroke Trial (IST) was one the biggest randomised trials in acute stroke. Methods: Available data on variables assessed at randomisation, at the early outcome point (14-days after randomisation or prior discharge) and at 6-months were extracted and made publically available. This data was collected over six years from 1991 to 1996. However the pilot phase was from 1991 to 1993. We will not be considering pilot phase since during that phase most of the variables were unavailable.

## Data Cleaning

Since we know that the pilot phase didn't have all data, we are going to remove data for pilot phase.

```
#Dropping null values for important variables
trial = trial.dropna(subset=['FDEAD','RATRIAL','FU2_DONE'])
```

```
#Removing unwanted columns as we have loads of variables
trial = trial.drop(columns = ['DHH14','DMAJNCHX','DSIDEX','DNOSTRKX','DNOSTRK','FDEADD',	
                              'DDEADX','FDEADX','HOURLOCAL','MINLOCAL','DAYLOCAL','FDEADC',
                              'DMAJNCHD','DMAJNCHX','SET14D','DRSHD','FLASTD','DPED',
                              'DSIDED','DRSISCD','DRSUNKD','NCCODE','DDEADD','DDEADC','ONDRUG'])
```

Now that we removed unwanted columns, let us create indicator variables for Stroke Types

```
#Convert Stroke types to binary data
dummy = pd.get_dummies(trial.STYPE)

trial = pd.concat([trial,dummy],axis= 1)
trial.head()
```

Large number of components have categorical values of 'C','N','Y' and 'U' for 'Can't Access', ' No','Yes' and 'Unknown' respectively. So let us replace all of them with numerical values.

```
#Converting character to numerical data 
cols = ["RSLEEP","RATRIAL","RCT","RVISINF","RHEP24", "RASP3","RDEF1","RDEF2","RDEF3","RDEF4","RDEF5","RDEF6","RDEF7","RDEF8","RXASP","DASP14","DASPLT","DLH14",
       "DMH14","DSCH","DIVH","DAP","DOAC","DGORM","DSTER","DCAA","DHAEMD","DCAREND","DTHROMB","DMAJNCH","DSIDE","DDIAGISC","DDIAGHA","DDIAGUN","DRSISC","DRSH",
       "DRSUNK","DPE","DALIVE","DDEAD","FDEAD","FRECOVER","FDENNIS","FAP","FOAC","CMPLASP","CMPLHEP"]
trial[cols] = trial[cols].replace({'C':0, 'N':1,'Y':2,'U':3})
```

As seen before the number of records where the survival status after 6 months is unknown is very small so we need to remove that

```
#Drop the columns having unknown status after 6 month follow up
trial = trial.drop(trial[trial.FDEAD == 3].index)
```

```
#Change null values for Discharge place and follow up place as "No Data" values
trial[['FPLACE', 'DPLACE']] = trial[['FPLACE','DPLACE']].fillna(value='N')
```

```
#Converting character to numerical data 
trial['RCONSC'] = trial['RCONSC'].replace({'D':0, 'F':1,'U':2})
trial['SEX'] = trial['SEX'].replace({'M':0, 'F':1})
trial['RXHEP'] = trial['RXHEP'].replace({'L':0,'N':1,'M':2})
```

```
#Inspecting distinct months
trial['RDATE'].unique()
```

We searched for these Randomization months in the internet and found that these are Polish name for months so we need to convert that to standard Gregorian calendar months.

```
#Convert Polish months to English months
dictionary = {'sty':'jan', 'lut':'feb', 'kwi':'apr', 'maj':'may', 'cze':'jun','lip':'jul',
                                         'sie':'aug', 'wrz':'sep', 'pa\x9f':'oct', 'lis':'nov', 'gru':'dec' } 
for key in dictionary.keys():
    trial['RDATE'] = trial['RDATE'].replace(key, dictionary[key],regex=True)
```

```
#Find top correlations and all the avaialable correlations
data_to_plot = trial80.drop(['DASP14', 'DASPLT', 'DLH14', 'DMH14', 'DSCH', 'DIVH', 'DAP', 'DOAC',
       'DGORM', 'DSTER', 'DCAA', 'DHAEMD', 'DCAREND', 'DTHROMB', 'DDIAGISC', 'COUNTRY',
       'DDIAGHA', 'DRSISC', 'DRSH', 'DRSUNK', 'DALIVE', 'DALIVED', 'DPLACE', 'RDATE',
       'DDEAD', 'FRECOVER', 'FDENNIS', 'FPLACE', 'FAP', 'FOAC','STYPE','FDEAD','Key'],1)

def get_redundant_pairs(data_to_plot):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = data_to_plot.columns
    for i in range(0, data_to_plot.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop
 
def get_top_abs_correlations(data_to_plot, n=5):
    au_corr = data_to_plot.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(data_to_plot)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Absolute Correlations")
print(get_top_abs_correlations(data_to_plot, 3))

#save top correlations to dataframe
corr_df_filter = pd.DataFrame(get_top_abs_correlations(data_to_plot, 100))
#make dataframe into 3 columns instead of 1 bc var names were treated as indices
corr_df_filter = corr_df_filter.reset_index()
#add column names that are easy to work with
corr_df_filter.columns = ['var1','var2', 'corr']
corr_df_filter[corr_df_filter['var1'] == 'FDEAD']

data_to_plot.corr()
```

## Supervised Modeling

Now we will try to model our data to provide predictions. We will be using PCA variables to predict our target. 

```
#Define test and train sets
X = trial80[['PCAC1','PCAC2','PCAC3']]
y = trial80[['FDEAD']]
offset = 2553

train = trial80[['PCAC1','PCAC2','PCAC3','FDEAD']][:offset]
test = trial80[['PCAC1','PCAC2','PCAC3']][offset+1:]
x_train, x_test = X[:offset], X[offset+1:] 
y_train, y_test = y[:offset], y[offset+1:] 
```

```
#Get the classes for target variable
le = LabelEncoder().fit(train.FDEAD) 
labels = le.transform(train.FDEAD)           
classes = list(le.classes_)  
```

```
#Function for evaluating confusion matrix
def conmatplot(var):
  plt.figure(figsize = (6,4))
  plt.clf()
  plt.imshow(var, interpolation='nearest', cmap='hsv')
  classNames = ['Not Dead','Dead']
  plt.title('Confusion Matrix - Test Data')
  plt.ylabel('True label')
  plt.xlabel('Predicted label')
  tick_marks = np.arange(len(classNames))
  plt.xticks(tick_marks, classNames)
  plt.yticks(tick_marks, classNames)
  s = [['TN','FP'], ['FN', 'TP']]

  for i in range(2):
      for j in range(2):
          plt.text(j,i, str(s[i][j])+"="+str(var[i][j]),horizontalalignment='center')
  plt.show()
```

```
#Try modelling using the 10 different classification models
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="rbf", C=0.025, probability=True),
    SGDClassifier(loss="log",penalty="l2", max_iter=5),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()]

# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
    clf.fit(x_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
       
    print('****Results****')
    y_pred = clf.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy: {:.4%}".format(acc))
    
    cv= cross_val_score(clf, x_train, y_train, cv=5)
    print("Cross validation scores: {}".format(cv))
    print("Average accuracy of cross validation: %0.2f (+/- %0.2f)" % (cv.mean(), cv.std() * 2))
    
    train_predictions = clf.predict_proba(x_test)
    ll = log_loss(y_test, train_predictions)
    print("Log Loss: {}".format(ll))
    
    skplt.metrics.plot_roc_curve(y_test, train_predictions)
    cm = confusion_matrix(y_test, y_pred)
    conmatplot(cm)
    
    log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
    log = log.append(log_entry)
    
print("="*30)
```
